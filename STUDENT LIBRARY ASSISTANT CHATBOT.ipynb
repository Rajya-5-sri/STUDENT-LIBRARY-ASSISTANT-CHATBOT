```python
books = {
    "python basics": {"available": 4, "fine_per_day": 2},
    "data science handbook": {"available": 2, "fine_per_day": 5},
    "machine learning": {"available": 1, "fine_per_day": 4},
    "dbms concepts": {"available": 3, "fine_per_day": 3}
}

student_records = {
    "101": {"name": "Aarav", "issued": 2, "max_limit": 3},
    "102": {"name": "Diya", "issued": 1, "max_limit": 3},
    "103": {"name": "Rahul", "issued": 0, "max_limit": 3}
}

WELCOME = """
===== Student and College Library Assistant Chatbot =====
Type one of these commands:
1. hello / hi
2. hours
3. books
4. search <book name>
5. issue <student_id> <book name>
6. return <student_id> <book name> <days late>
7. fine <student_id> <book name> <days late>
8. help
9. exit
"""

def show_books():
    print("\nAvailable books:")
    for title, info in books.items():
        print(f"- {title.title()} : {info['available']} copies")

def calculate_fine(book_name, late_days):
    return books[book_name]["fine_per_day"] * late_days

def chatbot_response(user_input):
    user_input = user_input.strip().lower()

    if user_input in ["hi", "hello", "hey"]:
        return "Hello! I am your College Library Assistant. How can I help you today?"

    elif user_input == "hours":
        return "Library timing is 8:00 AM to 8:00 PM from Monday to Saturday."

    elif user_input == "books":
        return "Type 'books' to view all available books."

    elif user_input == "help":
        return "You can search books, issue books, return books, and calculate fines."

    elif user_input.startswith("search "):
        book_name = user_input.replace("search ", "", 1)

        if book_name in books:
            copies = books[book_name]["available"]
            if copies > 0:
                return f"'{book_name.title()}' is available with {copies} copies."
            else:
                return f"'{book_name.title()}' is currently not available."
        else:
            return "Book not found in the library catalog."

    elif user_input.startswith("issue "):
        parts = user_input.split()

        if len(parts) < 3:
            return "Invalid format. Use: issue <student_id> <book name>"

        student_id = parts[1]
        book_name = " ".join(parts[2:])

        if student_id not in student_records:
            return "Student ID not found."
        elif book_name not in books:
            return "Book not found in the library catalog."
        elif books[book_name]["available"] <= 0:
            return "Sorry, this book is out of stock."
        elif student_records[student_id]["issued"] >= student_records[student_id]["max_limit"]:
            return f"Issue denied. {student_records[student_id]['name']} has reached the maximum book limit."
        else:
            books[book_name]["available"] -= 1
            student_records[student_id]["issued"] += 1
            return f"Book issued successfully to {student_records[student_id]['name']}."

    elif user_input.startswith("fine "):
        parts = user_input.split()

        if len(parts) < 4:
            return "Invalid format. Use: fine <student_id> <book name> <days late>"

        student_id = parts[1]

        try:
            late_days = int(parts[-1])
            book_name = " ".join(parts[2:-1])

            if student_id not in student_records:
                return "Student ID not found."
            elif book_name not in books:
                return "Book not found in the library catalog."
            elif late_days < 0:
                return "Late days cannot be negative."
            else:
                fine_amount = calculate_fine(book_name, late_days)
                return f"Fine for {student_records[student_id]['name']} is Rs. {fine_amount}."

        except ValueError:
            return "Invalid late days. Please enter a whole number."
        except Exception as e:
            return f"Unexpected error while calculating fine: {e}"

    elif user_input.startswith("return "):
        parts = user_input.split()

        if len(parts) < 4:
            return "Invalid format. Use: return <student_id> <book name> <days late>"

        student_id = parts[1]

        try:
            late_days = int(parts[-1])
            book_name = " ".join(parts[2:-1])

            if student_id not in student_records:
                return "Student ID not found."
            elif book_name not in books:
                return "Book not found in the library catalog."
            elif late_days < 0:
                return "Late days cannot be negative."
            elif student_records[student_id]["issued"] <= 0:
                return "No book is currently issued under this student account."
            else:
                books[book_name]["available"] += 1
                student_records[student_id]["issued"] -= 1

                if late_days > 0:
                    fine_amount = calculate_fine(book_name, late_days)
                    return f"Book returned successfully. Late fine is Rs. {fine_amount}."
                else:
                    return "Book returned successfully with no fine."

        except ValueError:
            return "Invalid late days. Please enter a whole number."
        except Exception as e:
            return f"Unexpected error while returning book: {e}"

    elif user_input == "exit":
        return "Thank you for using the College Library Assistant Chatbot."

    else:
        return "Sorry, I did not understand that command. Type 'help' to see options."

if __name__ == "__main__":
    print(WELCOME)

    while True:
        user_text = input("You: ")

        if user_text.strip().lower() == "books":
            show_books()
            continue

        reply = chatbot_response(user_text)
        print("Bot:", reply)

        if user_text.strip().lower() == "exit":
            break
            
```
